### P2_02.  
Write a program that reads an integer number from terminal, and shows a message only if the number is even.  
Example:  
~~~
Please, input a number: 4
Your number 4 is even
~~~


```python
# Your code here
n = eval(input('Please, input a number: '))
if n % 2 == 0:
    print('Your number {} is even'.format(n))
else:
    print('Your number {} is odd'.format(n))
```

### P2_03.  
Write a program that reads 3 numbers from terminal, **a, b y c** and tell us the biggest of the three.  
Example:  
~~~
Please, input a number: 4 
input a second number: 2
input a third one: 3
Biggest number is 4
~~~

```python
# Your code here
a = eval(input('Please, input a number: '))
b = eval(input('Please, input a number: '))
c = eval(input('Please, input a number: '))

if a >= b and a>=c:
    m = a
elif b >= a and b>=c:    
    m = b
else:
    m = c

print('Biggest number is {}'.format(m))
```
